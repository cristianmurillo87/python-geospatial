### Using matplotlib to plot a histogram of pixel values

```python
import numpy as np
import rasterio as rio
import math
import matplotlib.pyplot as plt
```

```python
img_file = 'data/20190321_174348_0f1a_3B_AnalyticMS.tif'
img = rio.open(img_file)
blue, green, red, nir  = img.read()
```

```python
fig  = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)

plt.title('Histogram Example', fontsize=18, fontweight='bold')
plt.xlabel('Pixel Values', fontsize=14)
plt.ylabel('Number of pixels', fontsize=14)

# discard nodata values
x = blue[np.not_equal(blue, img.nodata)]
bins = 20
color = 'lightgreen'

ax.hist(x, bins, color=color)
plt.show()
```
