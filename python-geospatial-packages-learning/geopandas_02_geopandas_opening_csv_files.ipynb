```python
# 1. Opening up the CSV with pandas
df = pd.read_csv('powerplants/powerplants.csv', sep=';')
# 2. Create a point using shapely by taking the lat/lon fields
points = df.apply(lambda row: Point(row.Longitude, row.Latitude), axis=1)
# 3. Original dataframe + geometry => geodataframe
powerplants = gpd.GeoDataFrame(df, geometry=points)
# 4. Indicate the geodataframe the CRS to be used
powerplants.crs = 'epsg:4326'
df.head()
```
